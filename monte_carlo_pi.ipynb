```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_pi(num_points):
    # Generate random points in the unit square
    x = np.random.uniform(0, 1, num_points)
    y = np.random.uniform(0, 1, num_points)

    # Calculate the distance of each point from the origin
    distances = np.sqrt(x**2 + y**2)

    # Count the number of points inside the quarter circle
    inside_circle = np.sum(distances <= 1.0)

    # Estimate pi
    pi_estimate = (inside_circle / num_points) * 4
    return pi_estimate, x, y, distances <= 1.0

# Number of points to simulate
num_points = 100000

# Run the Monte Carlo simulation
pi_estimate, x, y, inside_circle = monte_carlo_pi(num_points)
print(f"Estimated value of Ï€ using {num_points} points: {pi_estimate}")

# Visualization
fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(x[inside_circle], y[inside_circle], color='blue', s=1, label='Inside Circle')
ax.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1, label='Outside Circle')
circle = plt.Circle((0, 0), 1, color='black', fill=False, linewidth=2, linestyle='--')
ax.add_artist(circle)
ax.set_aspect('equal', 'box')
ax.legend(loc='best')
plt.show()
```
