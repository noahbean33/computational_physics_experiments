Simulation:

The Lorenz attractor is simulated using the simulate_lorenz function.
Plotting:

The trajectory of the Lorenz attractor is plotted in a 3D plot using Matplotlib.
This script should give you a nice visualization of the Lorenz attractor's chaotic behavior.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def lorenz_attractor(x, y, z, s=10, r=28, b=2.667):
    """
    Compute the next position of the Lorenz attractor.
    """
    x_dot = s * (y - x)
    y_dot = r * x - y - x * z
    z_dot = x * y - b * z
    return x_dot, y_dot, z_dot

def simulate_lorenz(x0, y0, z0, num_steps=10000, dt=0.01):
    """
    Simulate the Lorenz attractor starting from the initial conditions.
    """
    xs = np.empty(num_steps + 1)
    ys = np.empty(num_steps + 1)
    zs = np.empty(num_steps + 1)

    xs[0], ys[0], zs[0] = x0, y0, z0

    for i in range(num_steps):
        x_dot, y_dot, z_dot = lorenz_attractor(xs[i], ys[i], zs[i])
        xs[i + 1] = xs[i] + x_dot * dt
        ys[i + 1] = ys[i] + y_dot * dt
        zs[i + 1] = zs[i] + z_dot * dt

    return xs, ys, zs

# Initial conditions
x0, y0, z0 = 0.0, 1.0, 1.05

# Simulate the Lorenz attractor
xs, ys, zs = simulate_lorenz(x0, y0, z0)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```
